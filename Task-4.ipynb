## Task#4 - To Explore Decision Tree Algorithm

### For thre given 'iris' dataset, create the Decision Tree Classifier and Visualize it Graphically. The purpose is if we fee any new data to this classifier, it would be able to predict the right class accordingly.

### Load Datasets

```python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = datasets.load_iris()
```

```python
df=pd.DataFrame(data.data, columns=data.feature_names)
```

```python
x = df
x
```

```python
y = data.target
y
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x,y)
```

# plot Tree


```python
from sklearn import tree
```

```python
feature = data.feature_names
fig, axes = plt.subplots(figsize = (5,5), dpi=200)
tree.plot_tree(dtc,
               feature_names = feature,
               filled = True)
axes.set_title("Dicision Tree")
fig.savefig('imagename.png')
```
